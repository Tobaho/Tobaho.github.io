I am still making use of the Austin bike share trip data set. 

```python
%%bigquery --project tobahodataviz
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 100
```

```python
%%bigquery --project tobahodataviz

SELECT COUNT(*) as number_trips, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY number_trips DESC
```

```python
%%bigquery --project tobahodataviz

with t as (
  SELECT start_station_name AS Startingname, COUNT(start_station_name) AS numbertimes
  FROM bigquery-public-data.austin_bikeshare.bikeshare_trips
  GROUP BY Startingname
  ORDER BY numbertimes DESC
  LIMIT 2
)

SELECT Startingname, numbertimes
FROM t
ORDER BY numbertimes ASC
LIMIT 1
```

Riverside @S.Lamar has shown to be the second most popular station. 

```python
%%bigquery --project tobahodataviz

SELECT start_station_name, duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name
LIMIT 100
```

To calculate trips over an hour

```python
%%bigquery --project tobahodataviz

with t as (
SELECT start_station_name AS Trip, duration_minutes as DurationMin
FROM bigquery-public-data.austin_bikeshare.bikeshare_trips
)

SELECT COUNT(*) AS TripsOverHour
FROM t
WHERE DurationMin > 60
```

To calculate round trips

```python
%%bigquery --project tobahodataviz

with t as (
  SELECT start_station_name AS Trip, start_station_name AS Startingname, end_station_name AS Endingname
  FROM bigquery-public-data.austin_bikeshare.bikeshare_trips
)

SELECT COUNT(*) AS RoundTrips
FROM t
```
