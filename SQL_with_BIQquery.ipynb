```python
%%bigquery --project tobahodataviz
SELECT 
  max(duration_minutes)
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

I had a little bit of difficulty here. Just a tiny bit. Since I am making use of a different data set, I was confused on what to use as my average. Then i remembered 'Select' picks the column I want in the data table, and that's how I figured it out. I selected the column "duration_minutes. 
The question says "we should compute the longest trip from duration_minutes. I was not really sure so I just made use of "max(duration_minutes). 

```python
%%bigquery --project tobahodataviz
SELECT 
  AVG(duration_minutes)
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

```python
%%bigquery --project tobahodataviz
SELECT AVG(duration_minutes) as average_start_station_id
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` as table
```
