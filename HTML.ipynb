# Reading Data From the Web

For this project I will be making use of my favorite Tv show "Riverdale". 

```python
import requests
import pandas as pa
from bs4 import BeautifulSoup
```

This is the wikipedia page for Riverdale https://en.wikipedia.org/wiki/Riverdale_(2017_TV_series)

```python
r = requests.get('https://en.wikipedia.org/wiki/Riverdale_(2017_TV_series)')
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
#html_soup
```

I am going to display a title for the page

```python
html_soup.title
```

Now for retrieving all the links

```python
html_soup.table
```

```python
html_soup.table.find_all('a')
```

```python
links = html_soup.find_all('a', title=True)
RDURL = []
for link in links:
  RDURL.append('https://en.wikipedia.org/wiki/Riverdale_(2017_TV_series)' + link['href'])
print(RDURL)
```
