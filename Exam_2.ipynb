```python
df = df.rename(columns=lambda x: re.sub("\([^()]*\)", '', x))
df.columns
```

##1d

```python
df['activity_day'] = pa.to_datetime(df.activity_day, format = '%Y-%m-%d')
df.dtypes
```

##1e
Replacing the NaN's in the total_steps column

```python
MeanofStep = df['total_steps'].mean()
MeanofStep = math.trunc(MeanofStep)
df['total_steps'] = df['total_steps'].fillna(value = int(MeanofStep))
df.total_steps = df.total_steps.astype('int')

df['total_steps']
```

##1f 
Convert avg_pace into a float. To do this, i'll make use of fillna and mean. Just like the previous question 

```python
df['avg_pace'] = [re.sub(':','.', str(x)) 
for x in df['avg_pace']]
df.avg_pace = df.avg_pace.astype('float')
df['avg_pace']
df['avg_pace'] = df['avg_pace'].fillna(value = df['avg_pace'].mean())
df['avg_pace']
```
