##Integers and Float

I will start by importing my data sets

```python
import pandas as pa
import statistics as stat
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/iris.csv')
df.head().T
```

* Does converting SepalLength to integer increase or decrease the mean

```python
df.SepalLength.agg(['mean'])
```

```python

print("as int: ", stat.mean(df.SepalLength.astype('int')))
```

After converting, it decreases the mean.

* Does the direction of the shift remain the same if you groupby Class?







```python
df.SepalLength = df.SepalLength.astype('int')
df.groupby('Class').SepalLength.agg(['mean'])
```

From the above, I think the direction of shift remails the same

* Gather the mean, median, count and standard deviation of all columns when grouped by Class.

```python
df.groupby('Class').agg(['mean','median','count', 'std'])
```
