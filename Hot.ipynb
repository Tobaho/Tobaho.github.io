##One Hot Encoding

I will start this project by importing my dataset

```python
import pandas as pa

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')

df.head()
```

Convert the neighbourhood_group and the room_type into appropriate numerical data. Justify your decisisons on how you made the transformations.

I will start by breaking each row into individual pieces

```python
N = pa.crosstab(index=df['neighbourhood_group'], columns='count') 
N
```

Then i will make use of the get_dummies command to convert the column to a numerical data

```python
pa.get_dummies(df.neighbourhood_group).head()
```

I will use the above steps for the room_type column

```python
RT = pa.crosstab(index=df['room_type'], columns='count') 
RT
```

```python
pa.get_dummies(df.room_type).head()
```

I am trying to play around, so i'll transform the room_type column to  dictionary.. Let's see how that goes. 

```python
RT = sorted(df.room_type.unique())

RT
```

```python
eddict = {}

for i in range(3):
  eddict[RT[i]] = i


eddict
```

Putting it back in the data set

```python
df.room_type = df.room_type	.map(eddict)

df.head()
```

I changed my mind, I think I also want to try change the column neighbourhood group to a dictionary.

```python
NG = sorted(df.neighbourhood_group.unique())

NG
```

```python
eddict = {}

for i in range(5):
  eddict[NG[i]] = i


eddict
```

Putting it back in the data set. 

```python
df.neighbourhood_group = df.neighbourhood_group	.map(eddict)

df.head()
```
