#Scatterplot

To start off this project, i'll import the dataset

```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head()
```

Next, i'll create a scatterplot with many features

```python
ax = sns.scatterplot(data = df, x = 'calories', y = 'avg_heart_rate', hue = 'workout_type', style = 'max_heart_rate', size = 'avg_cadence', zorder = 2)
ax.set(title = 'Calories VS Maximum/ Average heart rate',
       xticks = [x for x in range(0,500,50)])
```

```python
ax = sns.scatterplot(data = df, x = 'calories', y = 'avg_heart_rate', hue = 'workout_type', style = 'max_heart_rate', size = 'avg_cadence', zorder = 2)
ax.set(title = 'Calories VS Maximum/ Average heart rate',
       xticks = [x for x in range(0,500,50)])
plt.legend(bbox_to_anchor=(1.25, 1), loc='upper left', borderaxespad=0)
sns.set(rc={'figure.figsize':(10,10)})

plt.axhspan(50, 100, facecolor= '0.8', alpha=0.3, zorder = 2)
plt.axhspan(100, 150, facecolor='0.6', alpha=0.3, zorder = 2)
plt.axhspan(150, 200, facecolor='0.4', alpha=0.3, zorder = 2)
plt.axhspan(200, 250, facecolor='0.2', alpha=0.3, zorder = 2)
plt.axhspan(250, 300, facecolor='0.2', alpha=0.3, zorder = 2)

plt.text(600, 100, "Light", horizontalalignment='left', size='medium', color='0.6', weight='semibold')
plt.text(600, 120, "Moderate", horizontalalignment='left', size='medium', color='0.8', weight='semibold')
plt.text(600, 160, "Anaerobic", horizontalalignment='left', size='medium', color='0.2', weight='semibold')
plt.text(600, 140, "Aerobic", horizontalalignment='left', size='medium', color='0.4', weight='semibold')
plt.text(600, 160, "Intensive", horizontalalignment='left', size='medium', color='0.2', weight='semibold')
plt.show()
```
