##Missing and Incomplete

I'll start by importing the dataset i'll be using

```python
import pandas as pa
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')

df.head()
```

To examine the number of nulls we have

```python
df.isnull().sum(axis = 0)
```

Next, I will impute at least two columns that are null in appropriate fashion. The two columns i reviewed are


*   last_review
*   reviews_per_month



```python
df[df.isnull().sum(axis = 1) == 0].last_review
```

```python
df.last_review.fillna(value = 0)
```

```python
df.reviews_per_month.fillna(value = 0)
```

```python
import numpy as np

df.last_review = df.last_review.fillna(df.groupby('minimum_nights').last_review.transform(lambda x: next(iter(x.mode()), np.nan)))

df
```

```python
df.fillna(df.groupby('minimum_nights').transform(lambda x: next(iter(x.mode()), np.nan)))
```
